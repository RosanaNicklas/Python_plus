```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({"X": [10, 20, 30, 40, 50], "Y":[15, 5, 10, 8, 6]})
df.head()
```

```python
df.plot()
```

```python
df.plot()
plt.show()
```

```python
df.plot(kind = "bar")
plt.show()
```

```python
df.plot(kind = "bar", stacked = True)
plt.show()
```

```python
df.plot(kind = "bar", stacked = True, grid = True, title ="Grafica ejemplo con pandas")
plt.show()
```

```python
df_2 = pd.DataFrame({"X": [10, 20, 30, 40, 50, 60, 70, 80],
"y": [10, 20, 10, 40, 20, 30, 20, 10]})
df_2
```

```python
df_2.y.value_counts().plot(kind = "bar")
plt.show()
```

```python
df_2.y.value_counts().plot(kind = "bar",
                        color = ["green", "orange"])
plt.show()
```

```python
df_2.y.value_counts().plot(kind = "bar",
                        color = ["green", "orange", "blue", "red"])
plt.show()
```
